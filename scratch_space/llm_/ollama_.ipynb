## Imports


```python
from time import time

from langchain_core.prompts.chat import (
    ChatPromptTemplate,
    HumanMessagePromptTemplate,
    SystemMessagePromptTemplate,
)
from langchain_ollama import ChatOllama
from loguru import logger as lg
```

## LangChain interface


```python
# define the model

llm = ChatOllama(
    model="llama3.1",
    temperature=0,
)
```

```python
# define the prompt

prompt = ChatPromptTemplate.from_messages(
    [
        SystemMessagePromptTemplate.from_template(
            "You are a helpful assistant that translates {input_language} to {output_language}."
        ),
        HumanMessagePromptTemplate.from_template("{input}"),
    ]
)
```

```python
# define the chain

chain = prompt | llm
```

```python
# chat with the agent

# start a timer
lg.info("Starting timer")
t0 = time()
res = chain.invoke(
    {
        "input_language": "English",
        "output_language": "German",
        "input": "I love programming.",
    }
)
t1 = time()
lg.info(res)
```

```python
# inspect the response and times

response_metadata = res.response_metadata
lg.info(response_metadata.keys())

lg.info(f"Time taken: {t1-t0:.6f} seconds")
lg.info(f"{response_metadata['total_duration']/1e9=}")
lg.info(f"{response_metadata['load_duration']/1e9=}")
lg.info(f"{response_metadata['prompt_eval_duration']/1e9=}")
lg.info(f"{response_metadata['eval_duration']/1e9=}")
```

## With external url


http://localhost:11434/api/chat


```python
ollama_url = ChatOllama(
    model="llama3.1",
    temperature=0,
    base_url="http://localhost:11434",
)
```

```python
# define the chain

chain_url = prompt | ollama_url
```

```python
# chat with the agent

# start a timer
lg.info("Starting timer")
t0 = time()
res = chain_url.invoke(
    {
        "input_language": "English",
        "output_language": "German",
        "input": "I love programming.",
    }
)
t1 = time()
lg.info(res)
```

```python
# inspect the response and times

response_metadata = res.response_metadata
lg.info(response_metadata.keys())

lg.info(f"Time taken: {t1-t0:.6f} seconds")
lg.info(f"{response_metadata['total_duration']/1e9=}")
lg.info(f"{response_metadata['load_duration']/1e9=}")
lg.info(f"{response_metadata['prompt_eval_duration']/1e9=}")
lg.info(f"{response_metadata['eval_duration']/1e9=}")
```
