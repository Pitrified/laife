## Setup


```python
from langchain_core.documents import Document
from laife.config.constants import VECTORSTORE_FOL
from laife.llm.vector_db import VectorDB
```

```python
documents = [
    Document(
        page_content="This is a test document",
        metadata={"title": "Test document 1"},
    ),
    Document(
        page_content="This is another test document",
        metadata={"title": "Test document 2"},
    ),
]
```

## Sentence Transformers


```python
from laife.config.lc_embeddings import (
    SentenceTransformerConfig,
    SentenceTransformerModelName,
)
```

```python
VECTORSTORE_FOL.mkdir(exist_ok=True)
```

```python
st_fol = VECTORSTORE_FOL / "sentence_transformers"
st_fol.mkdir(exist_ok=True)
```

```python
st_config = SentenceTransformerConfig(model_name=SentenceTransformerModelName.mpnet)

st_vdb = VectorDB(
    st_config,
    persist_directory=str(st_fol),
)
```

```python
st_vdb.add_documents(documents)
```

```python
st_vdb.add_documents(documents)
```

## OpenAI embeddings


```python
oai_fol = VECTORSTORE_FOL / "open_ai"
oai_fol.mkdir(exist_ok=True)
```

```python
from laife.config.credentials import OPENAI_API_KEY
from laife.config.lc_embeddings import OpenAIEmbeddingModelName, OpenAIEmbeddingsConfig
from laife.embed.utils import convert_to_secret_str

oai_config = OpenAIEmbeddingsConfig(
    OpenAIEmbeddingModelName.text_embedding_3_small,
    api_key=OPENAI_API_KEY,
)

oai_vdb = VectorDB(
    embeddings_config=oai_config,
    persist_directory=str(oai_fol),
)
```

```python
oai_vdb.add_documents(documents)
```

```python
oai_vdb.add_documents(documents)
```

### Do it better


```python
from langchain_openai import OpenAIEmbeddings

from laife.config.credentials import OPENAI_API_KEY

embedding_function_oai = OpenAIEmbeddings(
    api_key=OPENAI_API_KEY,
    model=OpenAIEmbeddingModelName.text_embedding_3_small.value,
)
oai_vdb_2 = VectorDB(
    embedding_function=embedding_function_oai,
    persist_directory=str(oai_fol),
)
```
