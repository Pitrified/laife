```python
class CustomOutput(BaseModel):
    """Custom output model for structured output."""

    summary: str = Field(..., description="A summary of the response")
    details: list[str] = Field(..., description="Detailed information about the response")


structured_model = default_model.with_structured_output(CustomOutput)
co = structured_model.invoke("Hello there!")
rprint(co)
```
